###Task 1 (2 pts)

1. Domain transformation: convert the above image (or any image of your choice) from spatial domain to frequency domain by using Fourier Transform. And display the **shifted** frequency FT image.
2. Circular Masking: Apply circle masks to preserve the center region and remove the rest area of the FT image. Use circles with radius of pixel size [1, 4, 8, 16, 32, 64, 128].
3. Image Reconstruction: reconstruct the images by performing Inverse Fourier Transform for all the masked FT image to return to spatial domain.

```python
# insert your code here

def convert_to_frequency(img):
    dft = np.fft.fft2(img)
    dft_shift = np.fft.fftshift(dft)
    return dft_shift

def apply_circular_mask(img,radius):
    rows, cols = img.shape
    mask = np.zeros((rows, cols), np.uint8)
    cv2.circle(mask, (cols//2, rows//2), radius, 1, -1)
    masked = convert_to_frequency(img) * mask
    return masked

def convert_back(freq_img):
    f_shift = np.fft.ifftshift(freq_img)
    img_back = np.fft.ifft2(f_shift)
    img_back = np.abs(img_back)
    return img_back



radii = [1, 4, 8, 16, 32, 64, 128]

plt.figure(figsize=(16, 6))

gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.subplot(2,5,1)
plt.imshow(gray_img,cmap='gray')
plt.title("Input Image")
plt.axis('off')

plt.subplot(2,5,2)
f_img = np.log(1+np.abs(convert_to_frequency(gray_img)))
plt.imshow(f_img, cmap='gray')
plt.title("Frequency Domain")
plt.axis('off')

for i, r in enumerate(radii):
    masked_freq = apply_circular_mask(gray_img, r)
    reconstructed = convert_back(masked_freq)

    plt.subplot(2, 5, i + 3)
    plt.imshow(reconstructed, cmap='gray')
    plt.title(f'Radius: {r}')
    plt.axis('off')

```

###Task 2 (2 pts)

The image Lenna_moire.png is contaminated with a periodic Moiré pattern. This effect usually appears in the frequency domain as concentrated "bursts" or "spikes" of high energy at specific coordinates that is far from the center of a shifted FT image. Please remove the moire effect by using Fourier Transform.

Note: use any masking methods you find appropriate (such as circular filtering, ring-shaped filtering, slicing out regions, etc.). The result might still have some artifacts (like faint ghosts of the pattern) even after filtering. Please experiment with different masking approaches and try your best to get a reasonable result.

```python
img_moire = cv2.imread("Lenna_moire.png")
img = cv2.cvtColor(img_moire, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
# insert your code here
rows, cols = img.shape
crow, ccol = rows//2, cols//2

f_shift = convert_to_frequency(img)
f_shift_copy = f_shift.copy()

dist = 50
size = 5

locations = [(-dist, 0), (dist, 0), (0, -dist), (0, dist)]


for r_off, c_off in locations:
    r_start, r_end = (crow + r_off) - size, (crow + r_off) + size
    c_start, c_end = (ccol + c_off) - size, (ccol + c_off) + size
    f_shift_copy[r_start:r_end, c_start:c_end] = 0

img_recon = convert_back(f_shift_copy)

plt.figure(figsize=(16, 6))
plt.subplot(1,4,1)
plt.title("Lenna Moire")
plt.imshow(img,cmap='gray')
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(img_recon,cmap='gray')
plt.title('Reconstruction')
plt.axis('off')

plt.subplot(1,4,3)
plt.title('Frequency Domain')
plt.imshow(np.log(1+np.abs(f_shift)),cmap='gray')
plt.axis('off')

plt.subplot(1,4,4)
plt.title('Masked Frequency')
plt.imshow(np.log(1+np.abs(f_shift_copy)),cmap='gray')
plt.axis('off')
```

```python
import math

# insert your code here
lenna = cv2.imread("Lenna.png",0)
cat = cv2.imread("cat.jpg",0)
cat = cv2.resize(cat, (lenna.shape[1], lenna.shape[0]))


def get_frequency_masked(freq,radius=25,center=0):
    rows, cols = freq.shape
    background = 1 - center
    mask = np.full((rows, cols), background, dtype=np.float32)

    cv2.circle(mask, (cols//2, rows//2), radius, center, -1)
    mask = cv2.GaussianBlur(mask, (51, 51), sigmaX=3)


    masked = freq * mask
    return masked



plt.figure(figsize=(16, 6))

plt.subplot(1,5,1)
plt.title("Lenna")
plt.imshow(lenna,cmap='gray')
plt.axis('off')

plt.subplot(1,5,2)
plt.title("Cat")
plt.imshow(cat,cmap='gray')
plt.axis('off')

cat_f = convert_to_frequency(cat)
lenna_f = convert_to_frequency(lenna)

plt.subplot(1,5,3)
plt.title("Lenna Low")
lenna_low = get_frequency_masked(lenna_f,center=1)
plt.imshow(convert_back(lenna_low),cmap='gray')
plt.axis('off')

plt.subplot(1,5,4)
plt.title("Cat High")
cat_high = get_frequency_masked(cat_f)
plt.imshow((convert_back(cat_high)),cmap='gray')
plt.axis('off')

plt.subplot(1,5,5)
plt.title("combined")
hybrid = lenna_low + cat_high
plt.imshow(convert_back(hybrid),cmap='gray')
plt.axis('off')
```

###Task 4-1 (4 pts) PCA

Implement Principal Component Analysis (PCA) using Singular Value Decomposition (SVD) to compress, visualize, and classify handwritten digits from the MNIST dataset. In the meantime, you need to:

1. Mean Image: Calculate and display the mean image;
2. Principal Components: Extract and visualize the 1st through 4th principal components (eigen-digits) to identify the primary directions of variance;
3. The starter code provides a test image randomly selected from the dataset (indices from sample_n to the end). Project this test image into the eigen-space. Find the closest point (best match/nearest neighbor) within the eigen-space to identify the digit. Reconstruct and display the image that is the closest point in the eigenspace.


```python
# Load MINIST handwritten digit dataset
# each image is 28x28 size (784), and numpy format
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
# load the first 5000 images as sample.
# Feel free to change sample size for better performance
sample_n = 5000
f_raw = torch.tensor(mnist.data.astype(np.float32))[:sample_n]
# ground truth labels of the samples
label = mnist.target.astype(int)[:sample_n]

print(f_raw.shape)
# Centering and Normalization
mean_image = torch.mean(f_raw, dim=0)
f_centered = f_raw - mean_image
f_normalized = f_centered / (torch.std(f_centered) + 1e-5)
f = f_normalized.t()
# PCA via SVD
U, S, Vh = torch.linalg.svd(f)

# insert your code here
plt.imshow(mean_image.reshape(28, 28), cmap='gray')
plt.title("Mean Image")
plt.axis('off')
plt.show()

fig, axes = plt.subplots(1, 4, figsize=(12, 3))
for i in range(4):
    eigen_digit = U[:, i].reshape(28, 28)
    axes[i].imshow(eigen_digit, cmap='gray')
    axes[i].axis('off')
    axes[i].set_title(f"PC {i+1}")
plt.show()


# Find k such that cumulative ratio > 0.95
# insert your code here

lambdas = S**2
total_var = torch.sum(lambdas)
var = lambdas / total_var
cumulative_var = torch.cumsum(var, dim=0)
k = torch.where(cumulative_var > 0.95)[0][0].item() + 1
print(f"Number of components for 95% variance: {k}")

# Project all the samples into the eigen space
# insert your code here
U_k = U[:, :k]
Omega = torch.mm(U_k.t(),f)

# get a test image with random index from sample_n to the end of the dataset
test_image_idx = np.random.randint(sample_n, len(mnist.data))
test_image_raw = torch.tensor(mnist.data.astype(np.float32))[test_image_idx].unsqueeze(0)
test_image_label = mnist.target.astype(int)[test_image_idx]

def get_image(test_image_raw):
    test_image_raw = torch.from_numpy(test_image_raw)
    test_f = (test_image_raw - mean_image) / (torch.std(f_centered) + 1e-5)
    test_omega = torch.mv(U_k.t(), test_f.flatten())
    distances = torch.norm(Omega - test_omega.unsqueeze(1), dim=0)
    closest_idx = torch.argmin(distances)
    return f_raw[closest_idx].reshape(28, 28)

# Project the test image into the eigen space
# Make sure you center and normalize it
test_f = (test_image_raw - mean_image) / (torch.std(f_centered) + 1e-5)
test_omega = torch.mv(U_k.t(), test_f.flatten())

# Find the nearest neighbor (best match/closet point)
distances = torch.norm(Omega - test_omega.unsqueeze(1), dim=0)
closest_idx = torch.argmin(distances)

# Reconstruct the nearest neighbor (best match/closet point)
omega_closest = Omega[:, closest_idx]
f_hat_norm = torch.mv(U_k, omega_closest)
nn_reconstructed = (f_hat_norm * (torch.std(f_centered) + 1e-5)) + mean_image

plt.figure(figsize=(10, 4))
plt.subplot(1, 3, 1)
plt.imshow(test_image_raw.reshape(28, 28), cmap='gray')
plt.title(f"Test Image (Label: {test_image_label})")

plt.subplot(1, 3, 2)
plt.imshow(f_raw[closest_idx].reshape(28, 28), cmap='gray')
plt.title(f"Nearest Neighbor (Label: {label[closest_idx]})")

plt.subplot(1, 3, 3)
plt.imshow(nn_reconstructed.reshape(28, 28), cmap='gray')
plt.title("NN Reconstructed")
plt.show()
```
